# Strategy

1) Put each day 1\$ of a stock

2) After a fixed time period sell the position

# Test

```python
df_test = pd.DataFrame(index=['timeperiod', 'EV', 'stdev', 'SR'])
```

```python
df_test
```

```python
def buy_hold_fixed(timeperiod):
    pocket_money = 0
    pocket_money_list = []
    count_day = 0
    for day in range(len(df_fb)-timeperiod):
        if not np.isnan(df_fb.values[day]):
            buy = df_fb.values[day]
            sell = df_fb.values[day+timeperiod]
            moneys = sell/buy - 1
            pocket_money_list.append(moneys)
            pocket_money += moneys
            count_day += 1
    
    SR = (pocket_money/(count_day/365)-0.02/(365-timeperiod))/np.std(pocket_money_list)
            
    # Store results
    df_test[str(1000+timeperiod)+'_days'] = [timeperiod, pocket_money/(count_day/365), np.std(pocket_money_list), SR]            
```

```python
for day in range(1, 365):
    buy_hold_fixed(day)
```

```python
final = df_test.transpose()
```

```python
final.head()
```

```python
df_test[final['SR'].idxmax()]
```

```python

# Set configuration of the graph
beingsaved =plt.figure(figsize=(18, 10))
plt.xlabel('Strategy', fontsize=20)
plt.ylabel('Returns', fontsize=20)
plt.ylim((0.00,100))
#plt.xlim((0.00,0.6))
mpl.rcParams['xtick.labelsize'] = 2 

## Define the labels
plt.title('Portfolios', fontsize=25)

plt.rc('font', size=16) 
plt.xticks(rotation=90)

# Create scatter plot coloured by Sharpe Ratio
plt.scatter(final.index,final.EV )#c=final.SR,cmap='RdYlBu'
#plt.colorbar()

plt.show()

beingsaved.savefig('SPX_EV.eps', format='eps', dpi=5000)
```
